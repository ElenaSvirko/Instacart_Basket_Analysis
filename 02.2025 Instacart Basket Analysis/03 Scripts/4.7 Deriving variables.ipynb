# 04 Deriving price filter column using user-defined function on the data subsection

```python
# defining the price_label funtion
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# creating the price_range column in the dataframe
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# checking the new column
df.head()
```

```python
# cheking value counts for the price_range column
df['price_range'].value_counts(dropna=False)
```

```python
# Cheking the max value in the price_range column
df['prices'].max()
```

# 05 Using the loc() Function to create new column in the data subsection

```python
# defining the conditions for the price_range column in the dataframe with the loc() function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna=False)
```

Same result was achieved using the user-defined if function and the loc function. Th loc function works faster, so is preferable for large dataframes.

# 07 Creating level of business column using for-loop 

```python
# Looking at how many orders are made during each week day
ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

```python
# Assigning the level of business to the 7 days of the week
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Looking at the business list
result
```

```python
# Defining the new 'busiest day' column
ords_prods_merge['busiest_day'] = result
```

```python
# Checking the values in the new column
ords_prods_merge['busiest_day'].value_counts(dropna=False)
```

# 09 Creating 'busiest_period_of_day' column - Exercise 4.7.4

```python
# Indentifying business during the day by hour
ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

The hours of the day will be split into equal thirds: "Most orders" = 10, 11, 14, 15, 13, 12, 16, and 9; "Fewest orders" = 23, 6, 0, 1, 5, 2, 4, and 3; "Average orders" = all remaining values in order_hour_of_day.

```python
# defining the three categories of business based on the number of orders during each hour of the day
result = []

for value in ords_prods_merge['order_hour_of_day']:
  if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value == 9:
    result.append('Most orders')
  elif value == 17 or value == 8 or value == 18 or value == 19 or value == 20 or value == 7 or value == 21 or value == 22:
    result.append('Average orders')
  else:
    result.append('Fewest orders')
```

```python
result
```

```python
# Defining the new 'busiest_period_of_day' column
ords_prods_merge['busiest_period_of_day'] = result
```

```python
# Checking the values in the 'busiest_period_of_day' column - Exercise 4.7.5
ords_prods_merge['busiest_period_of_day'].value_counts(dropna=False)
```
