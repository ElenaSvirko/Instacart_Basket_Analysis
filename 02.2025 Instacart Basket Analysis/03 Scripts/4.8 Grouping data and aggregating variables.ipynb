```python
# Descriptive statistics for prices of products purchased by each category of customer: loyal, regular and new - Exercise 4.8.5
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['min', 'max','mean', 'std', 'median']})
```

The products bought by the three loyalty groups have the same median price and very similar means. Thus loyal customers do not tend to buy more expensive products than regular or new customers. 

```python
# Creating 'mean_price' column with average product price for each user 
ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
# Checking that 'mean_price' column was defined correctly
ords_prods_merge.head(10)
```

```python
# Defining 'low spenders' and 'high spenders' based in the 'mean_price variable - Exercise 4.8.6
ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Checking the frequency of low and high spenders in this database
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
# Creating 'prior_order_days_median' column indicating the median number of days since prior order for each user
ords_prods_merge['prior_order_days_median'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
# Checking that the new column was created
ords_prods_merge.head()
```

```python
# Defining the frequentcustomer_flag - Exercise 4.8.7
ords_prods_merge.loc[ords_prods_merge['prior_order_days_median'] <= 10, 'frequentcustomer_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['prior_order_days_median'] > 10) & (ords_prods_merge['prior_order_days_median'] <= 20), 'frequentcustomer_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['prior_order_days_median'] > 20, 'frequentcustomer_flag'] = 'Non-frequent customer'
```

```python
# Checking the frequency of the categories of frequentcustomer_flag
ords_prods_merge['frequentcustomer_flag'].value_counts(dropna=False)
```

Some customers in the data base only made one order, so do not have anything in the 'days_since_prior_order' column (this was discussed during data cleaning). Thus there are 5 values that do not fit into the definitions for 'frequentcustomer_flag'. Customers with only one order can be counted as Non-frequent customers, so the definition can be adjusted accordingly.

```python
ords_prods_merge.loc[ords_prods_merge['prior_order_days_median'].isna(), 'frequentcustomer_flag'] = 'Non-frequent customer'
```

```python
# Rechecking the frequency of the categories of frequentcustomer_flag
ords_prods_merge['frequentcustomer_flag'].value_counts(dropna=False)
```

Customers with only one order have successfully been added to the Non-frequent customer category of 'frequentcustomer_flag'.
Exercise 4.8.7 completed.
